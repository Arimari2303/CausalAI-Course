#### BASIC MODEL

```r
library(sandwich)
```

```r
basic <- lwage ~ sex + exp1 + shs + hsg + mw + so + we + occ2+ ind2
regbasic.fit <- lm(basic, data=data)
#regbasic.fit # estimated coefficients
cat( "Number of regressors in the basic model:",length(regbasic.fit$coef), '\n')
regbasic.est <- summary(regbasic.fit)$coef[2,1]

HCV.coefs <- vcovHC(regbasic.fit, type = 'HC3');
regbasic.se <- sqrt(diag(HCV.coefs))[2] # Estimated std errors

confint(regbasic)[2,]
CI1upper <-confint(regbasic.fit)[2,2]
CI1low <-confint(regbasic.fit)[2,1]
```

#### FLEXIBLE MODEL

```r
flex <- lwage ~ sex + (exp1+exp2+exp3+exp4+shs+hsg+occ2+ind2+mw+so+we)**2
regflex.fit <- lm(flex, data=data)
regflex.est <- summary(regflex.fit)$coef[2,1]
#regflex.fit # estimated coefficients
cat( "Number of regressors in the flexible model:",length(regflex$coef)) # number of regressors in the Flexible Model

confint(regflex)[2,]
CI2upper <-confint(regflex.fit)[2,2]
CI2low <-confint(regflex.fit)[2,1]
```

##### Evaluating the performance of both models based on the (adjusted) $R^2_{sample}$ and the (adjusted) $MSE_{sample}$:

```r
# Assess the predictive performance

sumbasic <- summary(regbasic)
sumflex <- summary(regflex)

#  R-squared 
R2.1 <- sumbasic$r.squared
cat("R-squared for the basic model: ", R2.1, "\n")
R2.adj1 <- sumbasic$adj.r.squared
cat("adjusted R-squared for the basic model: ", R2.adj1, "\n")

R2.2 <- sumflex$r.squared
cat("R-squared for the flexible model: ", R2.2, "\n")
R2.adj2 <- sumflex$adj.r.squared
cat("adjusted R-squared for the flexible model: ", R2.adj2, "\n")
```

```r
# calculating the MSE
MSE1 <- mean(sumbasic$res^2)
cat("MSE for the basic model: ", MSE1, "\n")
p1 <- sumbasic$df[1] # number of regressors
MSE.adj1 <- (n/(n-p1))*MSE1
cat("adjusted MSE for the basic model: ", MSE.adj1, "\n")

MSE2 <-mean(sumflex$res^2)
cat("MSE for the flexible model: ", MSE2, "\n")
p2 <- sumflex$df[1]
MSE.adj2 <- (n/(n-p2))*MSE2
cat("adjusted MSE for the flexible model: ", MSE.adj2, "\n")
```

### Partialling out using Lasso - Case 1

```r
library(hdm)
library(sandwich)

# models
y <- lwage ~  exp1 + shs + hsg + mw + so + we + occ2+ ind2 # model for Y
d <- sex ~ exp1 + shs + hsg + mw + so + we + occ2+ ind2 # model for D

# partialling-out the linear effect of W from Y
t.Y <- rlasso(y, data=data)$res
# partialling-out the linear effect of W from D
t.D <- rlasso(d, data=data)$res

# regression of Y on D after partialling-out the effect of W
partial1.fit <- lm(t.Y~t.D)
partial1.est <- summary(partial1.fit)$coef[2,1]

cat("Coefficient for D via partialling-out", partial1.est)

# standard error
HCV.coefs <- vcovHC(partial1.fit, type = 'HC')
partial1.se <- sqrt(diag(HCV.coefs))[2]

# confidence interval
confint(partial1.fit)[2,]
CI1upper <-confint(partial1.fit)[2,2]
CI1low <-confint(partial1.fit)[2,1]
```

### Partialling out using Lasso - Case 2

```r
library(hdm)
library(sandwich)

# models
y <- lwage ~  (exp1+exp2+exp3+exp4+shs+hsg+occ2+ind2+mw+so+we)**2 # model for Y
d <- sex ~ (exp1+exp2+exp3+exp4+shs+hsg+occ2+ind2+mw+so+we)**2 # model for D

# partialling-out the linear effect of W from Y
t.Y <- rlasso(y, data=data)$res
# partialling-out the linear effect of W from D
t.D <- rlasso(d, data=data)$res

# regression of Y on D after partialling-out the effect of W
partial2.fit <- lm(t.Y~t.D)
partial2.est <- summary(partial2.fit)$coef[2,1]

cat("Coefficient for D via partialling-out", partial2.est)

# standard error
HCV.coefs <- vcovHC(partial2.fit, type = 'HC')
partial2.se <- sqrt(diag(HCV.coefs))[2]

# confidence interval
confint(partial2.fit)[2,]
CI2upper <-confint(partial2.fit)[2,2]
CI2low <-confint(partial2.fit)[2,1]
```

```r
'Plotting confidence intervals'
CIdf <- data.frame(x = c('Lasso1','Lasso2'),
                         y = c(regbasic.est,regflex.est),
                         lower = c(CI1low,CI2low),
                         upper = c(CI1upper,CI2upper))
```

```r
ggplot(CIdf, aes(x, y)) +        # ggplot2 plot with confidence intervals
  geom_point() +
  geom_errorbar(aes(ymin = lower, ymax = upper))

```
