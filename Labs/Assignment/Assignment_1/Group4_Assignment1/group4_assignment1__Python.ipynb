4.- ¬øCu√°ntas mujeres con t√≠tulo universitario o mayor tienen un salario correspondiente al 25% m√°s rico de la muestra? Mostrar el dataframe correspondiente

```python
#Definimos el 25% m√°s rico de la muestra
Mayor_riqueza = df['wage'].quantile(0.75)

#Definimos el grupo de mujeres, especificamos que deben tener un t√≠tulo universitario o superior
Mujeres_clg_ad = df.loc[(df['sex']==1) & ((df['clg']==1) | (df['ad']==1))]

#Del grupo de mujeres establecido, identificamos aquellas que pertenecen al cuantil m√°s rico.
Mujeres_ricas = Mujeres_clg_ad.loc[Mujeres_clg_ad['wage'] >= Mayor_riqueza]

#Empleamos el comando "len" para contar los datos filtrados de la lista deseada.
Num_mujeres_ricas = len(Mujeres_ricas)

#Presentamos los resultados
display(Mujeres_ricas)
print(f'Cantidad de mujeres con t√≠tulo universitario o superior cuyo salario corresponde al 25% m√°s rico de la muestra: {Num_mujeres_ricas}')
```

5.- ¬øCu√°ntos hombres con con grado completo secundario o inferior tienen un salario correspondiente al 25% m√°s rico de la muestra? Mostrar el dataframe correspondiente

```python
#Definimos el grupo de hombres, especificamos que deben tener grado completo secundario o inferior.
Hombres_shs_hsg = df.loc[(df['sex']==0) & ((df['shs']==1) | (df['hsg']==1))]

#Empleamos el grupo del cuantil m√°s rico previamente definido e identificamos los hombres que pertencen a este grupo
Hombres_ricos = Hombres_shs_hsg.loc[Hombres_shs_hsg['wage'] >= Mayor_riqueza]

#Empleamos el comando "len" para contar los datos filtrados de la lista deseada.
Num_hombres_ricos = len(Hombres_ricos)

#Presentamos los resultados
display(Hombres_ricos)
print(f'Cantidad de hombres con con grado completo secundario o inferior cuyo salario corresponde al 25% m√°s rico de la muestra: {Num_hombres_ricos}')
```

# 2. Wrangling de data

7.- Hacer un array para la variable Y, que ser√° el logaritmo del salario

```python
Y = np.array(LogSalario)
```

8.- Elaborar tres arrays nuevos para los predictores:

8.1. Modelo b√°sico

```python
# Retiramos las columnas que no son relevantes para la regresi√≥n o generan multicolinealidad.
df_clean = df_sin_salario.drop(columns=['rownames','shs','mw','occ','ind'])

#Generamos las variables dummy para la variable "occ2"
df_dummies_occ2 = pd.get_dummies(df_clean['occ2'], prefix='occ2', drop_first= True, dtype= int)

#Generamos las variables dummy para la variable "ind2"
df_dummies_ind2 = pd.get_dummies(df_clean['ind2'], prefix='ind2', drop_first= True, dtype= int)

#Unimos todos el dataframe principal con las dummies de ambas variables
Basic_model_1 = pd.concat([df_clean, df_dummies_occ2, df_dummies_ind2], axis=1)

#Eliminamos las columnas de las variables que convertimos a dummy del dataframe principal
Basic_model_0 = Basic_model_1.drop(columns= ['occ2','ind2'])

#Eliminamos las columnas de las variables de la variables que no se necesitan
Basic_model = Basic_model_0.drop(columns= ['exp2','exp3','exp4'])

#Convertir el modelo a un array de numpy
Basic_model_array = Basic_model.to_numpy()
```

8.2. Modelo flexible

```python
#Separamos la variable "sex" de las dem√°s para definir las interacciones. Las sumamos
df_interact_0 = Basic_model.drop(columns=['sex','exp1'])
df_interact_col = df_interact_0.columns

#Definimos un dataframe solo para la experiencia y su suma.
df_exp_0 = df_clean[['exp1','exp2','exp3','exp4']]
df_exp_columns = df_exp_0.columns

#Creamos el dataframe sobre el que se generar√° el bucle
Flex_model = pd.DataFrame()
```

```python
# Armamos el bucle para multiplicar todas las combinaciones de columnas
for col1 in df_interact_col:
    for col2 in df_exp_columns:
        nueva_columna = f'{col1}_{col2}'
        Flex_model[nueva_columna] = df_interact_0[col1] * df_exp_0[col2]

#Si sale un error de que la data est√° fragmentada, insistir. Solo se resiste.
```

```python
#Convertir el modelo a un array de numpy
Flex_model_array = Flex_model.to_numpy()
```

8.3 Modelo extra flexible

```python
#Apartamos la variable sex, que no va a tener interacci√≥n con las otras variables
df_sex = df['sex']

#Definimos un dataframe de las variables que s√≠ van a tener interacci√≥n
flex_interact = Basic_model_0.drop(columns=['sex'])
flex_col = flex_interact.columns
```

```python
#Definimos un bucle para la multiplicaci√≥n de las columnas
for i in range(len(flex_col)):
    for j in range(i + 1, len(flex_col)):
        col1 = flex_col [i]
        col2 = flex_col [j]
        nueva_columna = f'{col1}_{col2}'
        flex_interact[nueva_columna] = flex_interact[col1] * flex_interact[col2]
```

```python
#Unimos la variable que no interacciona con el conjunto de variables que s√≠ tuvieron interacci√≥n
Extraflex_model = pd.concat([df_sex, flex_interact], axis=1)

#Convertir el modelo a un array de numpy
Extraflex_model_array = Extraflex_model.to_numpy
```

```python
#Dividimos la data en test y train
Extraflex_model_array_train, Extraflex_model_array_test, Y_xf_train, Y_xf_test = train_test_split(Extraflex_model_array, Y)

#Planteamos el modelo extra flexible
Modelo_extra_flexible = LinearRegression()
Modelo_extra_flexible.fit(Extraflex_model_array, Y)

#Corremos el train
Y_xf_hat_train = Modelo_extra_flexible.predict(Extraflex_model_array_train)
display(Y_xf_hat_train)

#Corremos el test
Y_xf_hat_test = Modelo_extra_flexible.predict(Extraflex_model_array_test)
display(Y_xf_hat_test)
```
