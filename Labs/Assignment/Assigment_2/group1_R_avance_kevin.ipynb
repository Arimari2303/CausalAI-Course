4.Program a function that generates a logarithmically spaced grid. The input arguments should be the natural logarithms of the lower and upper bounds of the grid, as well as the natural logarithm of the spacing between each element of the grid. The output should be the logarithmically spaced grid, meaning that if we take the natural logarithm of each entry in the grid, they will be equally spaced. This will be the grid of values for  λ values to try during cross-validation.



```r
# Definir la función para generar una cuadrícula espaciada logarítmicamente
logarithmically_spaced_grid <- function(log_lower_bound, log_upper_bound, log_spacing) {
  # Calcular el número de puntos
  num_points <- ceiling((log_upper_bound - log_lower_bound) / log_spacing) + 1
  
  # Generar la cuadrícula espaciada logarítmicamente
  grid <- seq(log_lower_bound, log_upper_bound, length.out = num_points)
  
  # Convertir de logaritmos naturales a valores reales
  return(exp(grid))
}

# Ejemplo de uso
log_lower <- log(0.01)  # Logaritmo natural del límite inferior
log_upper <- log(100)   # Logaritmo natural del límite superior
log_spacing <- log(2)   # Logaritmo natural del espaciado

# Generar la cuadrícula
grid <- logarithmically_spaced_grid(log_lower, log_upper, log_spacing)

# Mostrar la cuadrícula generada
print(grid)
```

```r
# Tomar el logaritmo natural de los valores en la cuadrícula
log_grid <- log(grid)

# Verificar que los valores sean igualmente espaciados
diffs <- diff(log_grid)
print(diffs)
```

```r
# Definir la función para generar pliegues
generate_folds <- function(data_matrix, n_folds) {
  # Obtener el número de filas en la matriz de datos
  n_rows <- nrow(data_matrix)
  
  # Mezclar los índices de las filas
  set.seed(123)  # Para reproducibilidad
  indices <- sample(1:n_rows)
  
  # Dividir los índices en pliegues
  fold_indices <- split(indices, rep(1:n_folds, length.out = n_rows))
  
  return(fold_indices)  # Devolver la lista de pliegues
}

# Ejemplo de uso
data <- data.frame(x = rnorm(100), y = rnorm(100))  # Crear un data.frame de ejemplo
n_folds <- 5  # Definir el número de pliegues
folds <- generate_folds(data, n_folds)  # Generar los pliegues

# Mostrar los índices de los pliegues
print(folds)
```

```r
# Cargar las bibliotecas necesarias
library(glmnet)
Y <- data$wage  # Asegúrate de que 'wage' sea la variable de interés
X <- model.matrix(~ sex + (exp1 + exp2 + exp3 + exp4 + hsg + scl + clg + ad + so + we + ne + C(occ2) + C(ind2))^2 - 1, data = data)
X_normalized <- scale(X)  # Normalizar la matriz de predictores
# Definir la función
find_optimal_lambda <- function(Y, X, lambda_bounds, k) {
  # Crear la cuadrícula de lambda
  log_lower <- log(lambda_bounds[1])
  log_upper <- log(lambda_bounds[2])
  log_spacing <- log(2)  # Puedes ajustar esto según lo desees
  all_lambdas <- exp(seq(log_lower, log_upper, length.out = 100))
  
  # Generar pliegues
  folds <- generate_folds(data.frame(X), k)
  
  # Inicializar vectores para almacenar los MSE
  all_mse <- numeric(length(all_lambdas))
  
  # Ciclo sobre cada lambda
  for (i in seq_along(all_lambdas)) {
    lambda <- all_lambdas[i]
    fold_errors <- numeric(k)  # Almacenar los errores para cada pliegue
    
    # Ciclo sobre cada pliegue
    for (j in seq_along(folds)) {
      test_indices <- folds[[j]]
      train_indices <- unlist(folds[-j])
      
      # Dividir los datos en conjuntos de entrenamiento y prueba
      X_train <- X[train_indices, , drop = FALSE]
      y_train <- Y[train_indices]
      X_test <- X[test_indices, , drop = FALSE]
      y_test <- Y[test_indices]
      
      # Ajustar el modelo Lasso
      model <- glmnet(X_train, y_train, alpha = 1, lambda = lambda)
      
      # Predecir
      predictions <- predict(model, s = lambda, newx = X_test)
      
      # Calcular el MSE
      fold_errors[j] <- mean((y_test - predictions)^2)
    }
    
    # Calcular el MSE promedio para este lambda
    all_mse[i] <- mean(fold_errors)
  }
  
  # Encontrar el lambda óptimo
  optimal_index <- which.min(all_mse)
  optimal_lambda <- all_lambdas[optimal_index]
  
  # Ajustar el modelo final con el lambda óptimo para obtener los coeficientes
  final_model <- glmnet(X, Y, alpha = 1, lambda = optimal_lambda)
  optimal_coef <- as.vector(coef(final_model))
  
  # Devolver los resultados como una lista
  return(list(
    optimal_lambda = optimal_lambda,
    optimal_coef = optimal_coef,
    all_lambdas = all_lambdas,
    all_mse = all_mse
  ))
}

# Ejemplo de uso
# Suponiendo que ya tienes Y y X definidos
lambda_bounds <- c(0.01, 100)  # Ajusta según tus necesidades
k <- 5  # Número de pliegues

# Llamar a la función
results <- find_optimal_lambda(Y, X, lambda_bounds, k)

# Mostrar los resultados
print(results)
```
