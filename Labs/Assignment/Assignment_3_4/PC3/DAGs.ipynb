```python
import networkx as nx
import matplotlib.pyplot as plt


G = nx.DiGraph()

# Add nodes with labels
G.add_node('T', label='Smoking Behavior (T)')
G.add_node('Y', label='Lung Function (Y)')
G.add_node('A', label='Age (A)')
G.add_node('H', label='Height (H)')
G.add_node('S', label='Sex (S)')

# Add edges based on the DAG
edges = [
    ('S', 'T'),  # Sex → Smoking Behavior
    ('S', 'Y'),  # Sex → Lung Function
    ('A', 'T'),  # Age → Smoking Behavior
    ('A', 'H'),  # Age → Height
    ('A', 'Y'),  # Age → Lung Function
    ('H', 'Y'),  # Height → Lung Function
    ('T', 'Y')   # Smoking Behavior → Lung Function
]

G.add_edges_from(edges)

# Create a mapping of node labels
labels = nx.get_node_attributes(G, 'label')

# Define positions 
pos = {
    'A': (0, 2),
    'S': (0, 0),
    'T': (2, 1),
    'H': (2, 2),
    'Y': (4, 1)
}

# Draw the graph
plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=True, labels=labels, node_size=3000, node_color='lightblue', arrowsize=20)
plt.title('DAG for Youth Smoking and Lung Function')
plt.show()
```

```python
# Define the graph
G = nx.DiGraph()

# Add nodes with labels
G.add_node('T', label='Breastfed (T)')
G.add_node('Y', label='Number of Infections (Y)')
G.add_node('M', label='Marital Status (M)')
G.add_node('I', label='Family Income (I)')
G.add_node('E', label='Education (E)')
G.add_node('N', label='Number of Children (N)')
G.add_node('C', label='Childcare Outside Home (C)')

# Add edges based on the DAG
edges = [
    ('M', 'T'),  # Marital Status → Breastfed
    ('I', 'T'),  # Family Income → Breastfed
    ('E', 'T'),  # Education → Breastfed
    ('N', 'T'),  # Number of Children → Breastfed
    ('M', 'N'),  # Marital Status → Number of Children
    ('I', 'N'),  # Family Income → Number of Children
    ('E', 'N'),  # Education → Number of Children
    ('N', 'Y'),  # Number of Children → Number of Infections
    ('C', 'Y'),  # Childcare → Number of Infections
    ('M', 'C'),  # Marital Status → Childcare
    ('I', 'C'),  # Family Income → Childcare
    ('E', 'C'),  # Education → Childcare
    ('T', 'Y'),  # Breastfed → Number of Infections
    ('T', 'C')   # Breastfed → Childcare (if applicable)
]

G.add_edges_from(edges)

# Create a mapping of node labels
labels = nx.get_node_attributes(G, 'label')

# Define positions 
pos = {
    'M': (0, 2),
    'I': (0, 1),
    'E': (0, 0),
    'N': (2, 1),
    'T': (4, 1),
    'C': (4, 0),
    'Y': (6, 1)
}

# Draw the graph
plt.figure(figsize=(10, 6))
nx.draw(G, pos, with_labels=True, labels=labels, node_size=3000, node_color='lightgreen', arrowsize=20)
plt.title('DAG for Breastfeeding and Baby Infections')
plt.show()
```
