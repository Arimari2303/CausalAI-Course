### Bootstrapping and Confidence Intervals

1. **Generate 10,000 Bootstrap Estimates**:
   - Use a loop to generate 10,000 bootstrap estimates.


2. **Calculate 95% Confidence Intervals**:



```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.utils import resample

ols_model = LinearRegression()
ols_model.fit(X_train, y_train)
beta_hat = ols_model.coef_

n_iterations = 10000
bootstrapped_betas = []

for _ in range(n_iterations):
    X_sample, y_sample = resample(X_train, y_train)
    ols_model.fit(X_sample, y_sample)
    bootstrapped_betas.append(ols_model.coef_)

bootstrapped_betas = np.array(bootstrapped_betas)
```

Calculate the 95% confidence intervals

```python
beta_lower = np.percentile(bootstrapped_betas, 2.5, axis=0)
beta_upper = np.percentile(bootstrapped_betas, 97.5, axis=0)

for i, feature in enumerate(X.columns):
    print(f"{feature}: 95% CI [{beta_lower[i]}, {beta_upper[i]}]")
```

Fit a regression tree to predict the salary using all the features of your dataset. Follow these steps:

Calculate the out of sample mean squared error of the model. (2 pts)

```python
from sklearn.metrics import mean_squared_error

y_pred = ols_model.predict(X_test)
mse_ols = mean_squared_error(y_test, y_pred)
print(f"MSE OLS Model: {mse_ols}")
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import cross_val_score


tree_model = DecisionTreeRegressor(random_state=42)
tree_model.fit(X_train, y_train)
cross_val_scores = cross_val_score(tree_model, X_train, y_train, cv=5, scoring='neg_mean_squared_error')
print(f"MSE Cross-Validated Tree Model: {-np.mean(cross_val_scores)}")
y_pred_tree = tree_model.predict(X_test)
mse_tree = mean_squared_error(y_test, y_pred_tree)
print(f"MSE Decision Tree Model: {mse_tree}")
```

Which model performs better in terms of predictive accuracy? (2 pts)

```python
if mse_ols < mse_tree:
    print("El modelo OLS tiene mejor precisión predictiva.")
else:
    print("El modelo de Árbol de Decisión tiene mejor precisión predictiva.")
```
