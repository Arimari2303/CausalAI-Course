# 2. Creating Lasso Cross-Validation Procedure 

4. 

4.1 Emake function 

```r
emake <- function(){

    options <- ("warn"=-1)

    e <- 0 

    for (n in 0:10000){

        e <- e + 1/(factorial(n))
    }
    return(e)
}
```

4.2

```r
log_space_grid <- function (lower_bound, upper_bound, space_grid) {

    initial_seq <- seq(lower_bound, upper_bound, length.out = space_grid)

    for (i in initial_seq){
        
        log_sequence <- c()

        log_ind <- print(i^emake())

        log_sequence <- c(log_sequence, log_ind)
    }
    return(log_sequence)
}
```

```r
log_space_grid <- log_space_grid(0,1,1000)
log_space_grid
```

```r
kfolds <- function(df, ngroups){

  num_folds <- split.data.frame(df, floor(seq(1, ngroups, length.out = nrow(df))))

  return(num_folds)
}
```

```r
folds_x <- kfolds(flex_matrix, 5)
```

```r
cross_validation_procedure <- function(y, x, lower_bound, upper_bound, nfolds){

    initial_seq<- seq(lower_bound, upper_bound, length.out = 1000)

    log_seq <- c()

    for (i in initial_seq){
        
        log_indi <- print(i^emake())

        log_seq <- c(log_seq,log_indi)
    }

    final_list <- c()
  folds <- kfolds(x, nfolds)

  av_mse <- c()

    for (logi in log_seq){

        mean_sq_perfold <- c()

        for (train_index in folds){

        x_train <- x[train_index,]
        y_train <- y[train_index]
    
        test_index <- setdiff(seq_len(nrow(x)), train_index)
         x_test <- x[test_index,]
         y_test <- y[test_index]
        
        # Fit lasso_model
        model <- glmnet(x_train, y_train, alpha = 1, lambda = logi)

        #Predict 

        predict <- predict(model, x_test)

        prediction_error <- sum((y_test - predict)^2)

        mean_sq_perfold <- c(mean_sq_perfold, prediction_error)
        }
        av_mse <- c(av_mse, mean(mean_sq_perfold))
    }
    
    
    chosen_alpha <- log_seq[which.min(av_mse)]
    
    best_model <- glmnet(x_train, y_train, alpha = 1, lambda = chosen_alpha)

    best_model_coeffs <- coef(best_model)[2:length(coef(best_model))]

    final_list <- c(final_list, chosen_alpha, best_model_coeffs, log_seq, av_mse )

    return(final_list)
}
```
