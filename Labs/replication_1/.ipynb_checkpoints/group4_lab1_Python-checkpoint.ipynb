# 3. Replication graphic paper

```python
a = data[["exp1","sex","lwage","exp2","exp3","exp4","scl","clg"]]
male = data[ data[ 'sex' ] == 0 ]
d_male = male[ [ "exp1","sex","lwage","exp2","exp3","exp4","scl","clg" ] ]
d_male.shape
```

```python
pip install statsmodels
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
y = 'lwage ~  (exp1+exp2+exp3+exp4)**4'
t_y = smf.ols( formula = y , data = d_male )
ty_predict = t_y.fit().predict()
ty_predict.tolist()
d_male['predict'] = ty_predict
d_male
```

```python
d_male_scl = d_male[d_male['scl'] == 1]
d_male_clg = d_male[d_male['clg'] == 1]
```

```python
scl = d_male_scl[['lwage','exp1','predict']].groupby(['exp1']).mean().reset_index()
scl['promed'] = 0
scl['promed_fit'] = 0
scl.shape
```

```python
clg = d_male_clg[['lwage','exp1','predict']].groupby(['exp1']).mean().reset_index()
clg['promed'] = 0
clg['promed_fit'] = 0
clg.shape
```

```python
for i in range(0,79):
    aaa = d_male_scl[d_male_scl['exp1'] <= scl['exp1'][i]]
    scl['promed'][i] = aaa['lwage'].mean()
    scl['promed_fit'][i] = aaa['predict'].mean()
print(scl['promed'],scl['promed_fit'])
```

```python
for i in range(0,39):
    aaa = d_male_clg[d_male_clg['exp1'] <= clg['exp1'][i]]
    clg['promed'][i] = aaa['lwage'].mean()
    clg['promed_fit'][i] = aaa['predict'].mean()
print(clg['promed'],clg['promed_fit'])
```

```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams["figure.figsize"] = (20,10)
plt.rcParams.update({'font.size': 15})
x = clg["exp1"]
y = clg['promed']
x_1 = scl['exp1']
y_1 = scl['promed']
x_2 = clg['exp1']
y_2 = clg['promed_fit']
x_3 = scl['exp1']
y_3 = scl['promed_fit']

plt.plot(x, y, label = "Collage Graduated - Actual", linestyle='-')    
plt.plot(x_1, y_1, label = "Some College Education - Actual", linestyle='-')
plt.plot(x_2, y_2, label = "Collage Graduated - Fitted", linestyle='dashed')
plt.plot(x_3, y_3, label = "Some College Education - Fitted", linestyle='dashed')

plt.xticks()
plt.legend()
plt.title("Wage for mens by college graduated and some college")
plt.xlabel("Years of Potencial Experience")
plt.ylabel("Log Wage(or Wage Gap)")
plt.show()
```
