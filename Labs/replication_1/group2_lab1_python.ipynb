### Unconditional regression

```python
#An unconditional comparision is generally biased. Let's make some regressions to apply some Econometrics to capture the causal effect of college education in wages.

#install pkgs
!pip install statsmodels

import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
#previously clean our dataset (some college v. college completed subset)
subset= data.copy()
subset
```

```python
subset['clg_comp'] = np.where(subset['clg']==1, 1, 0)

subset
```

```python
#dispense with individual with no completed nor some college education
subset = subset[subset['shs']==0]
subset = subset[subset['hsg']==0]
subset = subset[subset['ad']==0]

#dispense with clg and scl columns
subset.drop(['shs','hsg','ad'], axis='columns', inplace=True)

subset
```

#### Make regressions to find the causal effect of college education in wages.

```python
#remember: ~ = Alt + 126

nocontrol_model = smf.ols(formula = 'lwage ~ clg_comp', data=subset )
nocontrol_est = nocontrol_model.fit().summary2().tables[1]['Coef.']['clg_comp']
nocontrol_est
nocontrol_se2 = nocontrol_model.fit().summary2().tables[1]['Std.Err.']['clg_comp']

# robust standar erros
HCV_coefs = nocontrol_model.fit().cov_HC0
nocontrol_se = np.power( HCV_coefs.diagonal() , 0.5)[1]
nocontrol_se

# print unconditional effect of gender and the corresponding standard error

print( f'The estimated college coefficient is {nocontrol_est} and the corresponding standard error is {nocontrol_se2}' )
print( f'The estimated college coefficient is {nocontrol_est} and the corresponding robust standard error is {nocontrol_se}','\n' )
```

We obtain the unconditional college wage gap $28,63 \%$ is the same as the no control OLS regression estimator, $0.2863$
