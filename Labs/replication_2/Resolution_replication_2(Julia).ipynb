## Prediction Question

```julia
#basic model
basic  = @formula(lwage ~ (sex + exp1 + shs + hsg + mw + so + we + occ2+ ind2))
basic_results  = lm(basic, data)
```

```julia
beta_basic = GLM.coeftable(basic_results).cols[1][2]
beta_basic_se = GLM.coeftable(basic_results).cols[2][2]
println("Coefficient for basic model " , beta_basic, "standard error:", beta_basic_se)
```

```julia
# couples variables combinations 
combinations_upto(x, n) = Iterators.flatten(combinations(x, i) for i in 1:n)

# combinations without same couple
expand_exp(args, deg::ConstantTerm) =
    tuple(((&)(terms...) for terms in combinations_upto(args, deg.n))...)

StatsModels.apply_schema(t::FunctionTerm{typeof(^)}, sch::StatsModels.Schema, ctx::Type) =
    apply_schema.(expand_exp(t.args_parsed...), Ref(sch), ctx)
```

```julia
# Flexible model
    
flex_model = @formula(lwage ~ sex + (exp1+exp2+exp3+exp4+shs+hsg+occ2+ind2+mw+so+we)^2)
flex_model = apply_schema(flex_model, schema(flex_model, data))
```

```julia
flex_results = lm(flex_model, data)
```

```julia
beta_flexible = GLM.coeftable(flex_results).cols[1][2]
beta_flexible_se = GLM.coeftable(flex_results).cols[2][2]
```

```julia
println("Coefficient for flexible model " , beta_flexible, "standard error:", beta_flexible_se)
```

### Assess the predictive performance


```julia
# R squared
r2_basic = adjr2(basic_results)
r2_flexible = adjr2(flex_results)

# R squared_ajusted
r2adj_basic = adjr2(basic_results)
r2adj_flexible = adjr2(flex_results)
```

```julia
y = data[!,:lwage]

#Prediction 

y_predict_basic = predict(basic_results)
y_predict_flex = predict(flex_results)
```

```julia
# MSE
mse_basic = mse(y_predict_basic, y)
mse_flexible = mse(y_predict_flex, y)

n = length(y)  # total observations
p1 = length(GLM.coef(basic_results)) # number of regresors
p2 = length(GLM.coef(flex_results)) # number of regresors

# R MSE_ajusted
mseadj_basic = n/(n-p1)*mse(y_predict_basic, y)
mseadj_flexible = n/(n-p2)*mse(y_predict_flex, y)
```
